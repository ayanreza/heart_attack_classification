```python
# Separate Data into features and labels

X = df.drop(['output'], axis=1).values
y = df[['output']].values
```

```python
print(X)
```

```python
print(y)       
```

```python
# Divide data into test and training sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
# Perform standardization

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Fit all the models to the training data

from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
import seaborn as sns


knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train, y_train.ravel())

gnb = GaussianNB()
gnb.fit(X_train, y_train.ravel())

linear = svm.SVC(kernel='linear', C=1, decision_function_shape='ovo').fit(X_train, y_train.ravel())
rbf = svm.SVC(kernel='rbf', gamma=1, C=1, decision_function_shape='ovo').fit(X_train, y_train.ravel())
poly = svm.SVC(kernel='poly', degree=3, C=1, decision_function_shape='ovo').fit(X_train, y_train.ravel())
sig = svm.SVC(kernel='sigmoid', C=1, decision_function_shape='ovo', probability=True).fit(X_train, y_train.ravel())
 
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

rf = RandomForestClassifier(n_estimators=10, criterion='gini',random_state=1)
rf.fit(X_train,y_train.ravel())
```

```python
# 5 Fold setup 

kf = KFold(n_splits=5)
```

```python
# Use for loop to dynamically go over all models and save cross val results
models = [knn, gnb, linear, rbf, poly, sig, dt, rf]

cross_val_results = []

for model in models:
    result = cross_val_score(model, X_train, y_train.ravel(), cv = kf, scoring='accuracy')
    cross_val_results.append(result)
```

```python
# Print the average accuracy for each model according too cross validation 

for i, model in enumerate(models):
    print(f"{model} Cross-Validation Results:")
    print(cross_val_results[i])
    print(f'Average Accuracy: {cross_val_results[i].mean()}')
    print()
```

```python
# Plot average accuracy for each model

for i, model in enumerate(models):
    plt.plot(range(1, len(cross_val_results[i]) + 1), cross_val_results[i], marker='o', linestyle='-', label=model)
```

```python
# Creates classifcation report for all models and stores in dictionary

reports = {}

for model in models:
    prediction = model.predict(X_test)
    class_report = classification_report(y_test, prediction, output_dict=True)
    reports[str(model)] = class_report

print(reports)
```

```python
# Extra graph to check what is the most optimal K value 

from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

accuracy = []

for i in range(1,11):
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train, y_train.ravel())
  y_pred = knn.predict(X_test)
  accuracy.append(accuracy_score(y_test, y_pred.ravel()))
plt.plot(list(range(1,11)), accuracy, color='blue', linestyle='dashed', marker='x')
plt.xlabel('k value')
plt.ylabel('accuracy')
plt.title('k value vs accuracy')
plt.show()
```

```python
#Graph showing congregated classifcation report data for each model 

#models = [knn, gnb, linear, rbf, poly, sig, dt, rf]

accuracy = []
precision = []
recall = []
f1 = []

for model in models:
    report = reports[str(model)]

    accuracy.append(round(report['accuracy'],2))
    precision.append(round(report['weighted avg']['precision'],2))
    recall.append(round(report['weighted avg']['recall'],2))
    f1.append(round(report['weighted avg']['recall'],2))

print(accuracy)
print(precision)
print(recall)
print(f1)

# Number of bars
n_bars = len(models)

# Number of groups of bars (accuracy, precision, recall, f1)
n_groups = 4

# Total width for each group of bars
total_width = 0.8

# Width for each individual bar within a group
bar_width = total_width / n_groups

# Calculate the center positions for each group of bars
X_axis = np.arange(len(models))

# Calculate the left positions for each individual bar within a group
left_positions = X_axis - (total_width - bar_width) / 2

# Setting the size of the graph
plt.rcParams["figure.figsize"] = (50, 15)

# Plotting the bars for each metric using the left positions and respective data
plt.bar(left_positions, accuracy, bar_width, label='Accuracy')
plt.bar(left_positions + bar_width, precision, bar_width, label='Precision')
plt.bar(left_positions + bar_width * 2, recall, bar_width, label='Recall')
plt.bar(left_positions + bar_width * 3, f1, bar_width, label='F1')

# Setting the range of values for the X axis and labels
plt.xticks(X_axis, models)
plt.xlabel("Model")
plt.ylabel("Weighted Avg")
plt.title("Model vs Weighted Avg Of Various Metrics")
plt.legend(loc='upper right')

plt.show()
```
